```python
import pandas as pd

df = pd.read_csv('NYCinspections.csv')
pd.set_option('display.max_columns', 27)
df.columns
```

```python
filtered_df = df[['CAMIS', 'DBA', 'BORO', 'BUILDING', 'STREET', 'ZIPCODE', 'CUISINE DESCRIPTION', 
                  'INSPECTION DATE', 'ACTION', 'VIOLATION CODE', 'VIOLATION DESCRIPTION', 'CRITICAL FLAG', 
                  'SCORE', 'GRADE', 'INSPECTION TYPE', 'Latitude', 'Longitude']]
filtered_df
```

```python
date_df = filtered_df[['CAMIS', 'INSPECTION DATE']].copy()

date_df['INSPECTION DATE'] = pd.to_datetime(date_df['INSPECTION DATE']).dt.year

date_df.groupby('INSPECTION DATE').count()
```

```python
date_filtered_df = filtered_df

date_filtered_df['INSPECTION DATE'] = pd.to_datetime(filtered_df['INSPECTION DATE'])

newdf = date_filtered_df[date_filtered_df['INSPECTION DATE'] > '1/1/2010']

newdf

#Checking to make sure the 1900s are filtered out

date_dfl = newdf[['CAMIS', 'INSPECTION DATE']]

#date_dfl['INSPECTION DATE'] = date_dfl['INSPECTION DATE'].dt.month
date_dfl['INSPECTION DATE'] = date_dfl['INSPECTION DATE'].dt.year

date_dfl.groupby('INSPECTION DATE').count()
```

```python
newdf.groupby(['INSPECTION TYPE'])[['SCORE', 'GRADE']].count()
```

```python
cycle_inspection_df = newdf[newdf['INSPECTION TYPE'].isin(['Cycle Inspection / Initial Inspection', 'Cycle Inspection / Re-inspection'])]

cycle_inspection_df.to_excel('ijexcel.xlsx')
```

```python
violation_df = cycle_inspection_df[['VIOLATION CODE', 'VIOLATION DESCRIPTION']].drop_duplicates(subset='VIOLATION CODE')

violation_df.sort_values('VIOLATION CODE').to_excel('ViolationCSV.xlsx')

```
